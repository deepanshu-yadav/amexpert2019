```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
original_df = pd.read_csv('data.csv')

mask = ~original_df['is_goal'].isnull()
#not_mask = original_df['is_goal'].isnull()

full_list_of_columns = original_df.columns

def get_numerical_columns(list_of_cols , df):
    new_list = []
    for col in list_of_cols:
        if df[col].dtype != 'object':
            new_list.append(col)
            
    return new_list

num_columns = get_numerical_columns(full_list_of_columns , original_df)

#num_df = original_df.loc[ : , num_columns]

#clean_num_df = num_df.drop(  ['Unnamed: 0' , 'match_event_id'  , 'shot_id_number' , 'match_id' , 'team_id' ] , axis =1 )
#clean_num_df.head()

#df_cr = clean_num_df.corr(method='pearson')


original_df["area_of_shot"] = original_df["area_of_shot"].astype('category')

original_df["area_of_shot_conv"] = original_df["area_of_shot"].cat.codes

original_df["type_of_shot"] = original_df["type_of_shot"].astype('category')

original_df["type_of_shot_conv"] = original_df["type_of_shot"].cat.codes


original_df["shot_basics"] = original_df["shot_basics"].astype('category')

original_df["shot_basics_conv"] = original_df["shot_basics"].cat.codes


num_columns_new = num_columns + [ 'type_of_shot_conv' , 'shot_basics_conv' , 'area_of_shot_conv'  ]


#mask = ~original_df['is_goal'].isnull()

#print(len(mask))




clean_df= original_df.loc[: ,  num_columns_new ]
clean_df = clean_df.drop(  ['Unnamed: 0' , 'match_event_id'  , 'shot_id_number' , 'match_id' , 'team_id' ] , axis =1 )
#clean_num_df.head()

#clean_df.head()

#print(  clean_df['is_goal'].isnull().count())



#X_full = clean_df[~clean_df['is_goal'].isnull()].drop(['is_goal'] , axis=1)
#Y_full = clean_df[~clean_df['is_goal'].isnull()].is_goal

X_train = clean_df[mask]
Y_train = clean_df['is_goal'][mask]

X_test = clean_df[~mask]






#print(mask)
```

```python
#X_test

import xgboost as xgb
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np


xg_reg = xgb.XGBRegressor(objective ='multi:softprob', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10 , num_class =2)


xg_reg.fit(X_train,Y_train)

preds = xg_reg.predict(X_test)
```
